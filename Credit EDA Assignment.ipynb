```python
#Calculating percentage of series of columns
def percentage(series):
    Total_sum = series.isnull().sum()
    Length_of_series = len(series)
    return round(Total_sum/Length_of_series, 4)
percentage(df)
```

```python
#Dropping columns having more than 20% null values in the dataset
for col, values in df.iteritems():
    if percentage(df[col]) > 0.19: #took 0.19 to remove EXT_SOURCE_3 specially as it is same as EXT_SOURCE_2 and 3 as normalized score from external data as it looks applicable for this case only but in general we can seperately drop it too
        df.drop(col, axis=1, inplace=True)
df.head()
```

```python
#Now we have dataset columns with negligible null data but we have to fill empty null values by using .fillna()
#First we have to check for category of column each one by one then fill it accordingly

df.NAME_TYPE_SUITE.value_counts()
```

```python
#Since Unaccompained data has the highest number in counts, we can use mode to fill the missing values
df["NAME_TYPE_SUITE"].fillna(df["NAME_TYPE_SUITE"].mode(),inplace=True)
```

```python
#Same as above here also values difference is low we can fill it with mean values
df['EXT_SOURCE_2'].fillna((df['EXT_SOURCE_2'].mean()), inplace=True)
```

```python
#Previous_data interpretation
pre_data = pd.read_csv("previous_application.csv") 
pre_data.head()
```

```python
pre_data.info()
```

```python
pre_data.isnull().sum()
```

```python
D =((pre_data.isnull().sum())*100/len(pre_data)) #percentage of null values with respect to length of dataset(for each column)
D
```

```python
#Calculating percentage in the series
def percentage(series):
    Total_prev_sum = series.isnull().sum()
    series_length = len(series)
    return round(Total_prev_sum/series_length, 3)
percentage(pre_data)
```

```python
#Now we drop columns with more than 20% null values
for col, values in pre_data.iteritems():
    if percentage(pre_data[col]) > 0.20: 
        pre_data.drop(col, axis=1, inplace=True)
pre_data.head()
```

```python
list(pre_data.columns)
```

```python
pre_data= pre_data.filter(['SK_ID_PREV',
 'SK_ID_CURR',
 'NAME_CONTRACT_TYPE',
 'AMT_APPLICATION',
 'AMT_CREDIT',
 'WEEKDAY_APPR_PROCESS_START',
 'HOUR_APPR_PROCESS_START',
 'FLAG_LAST_APPL_PER_CONTRACT',
 'NFLAG_LAST_APPL_IN_DAY',
 'NAME_CLIENT_TYPE',
 'NAME_CASH_LOAN_PURPOSE',
 'NAME_CONTRACT_STATUS',
 'DAYS_DECISION',
'SELLERPLACE_AREA'])
```

```python
#Visualizing our analysis for better understanding
app_status = pre_data['NAME_CONTRACT_STATUS']
app_status.head()
```

```python
app_contract=round((app_status.value_counts()/pre_data["NAME_CONTRACT_STATUS"].count())*100,2)
app_contract = pd.DataFrame(app_contract)
app_contract.reset_index(level=0, inplace=True)
app_contract.rename(columns=  {"index": "NAME_CONTRACT_STATUS", 
                     "NAME_CONTRACT_STATUS":"Percentage"}, 
                                 inplace = True) 
app_contract.sort_values(by = 'Percentage' , inplace = True, ascending = False)
app_contract
```

```python
plt.bar(app_contract['NAME_CONTRACT_STATUS'],app_contract['Percentage'],color=['green', 'red', 'blue', 'cyan'])
plt.show()
```

```python
#Checking how many repeaters of loan debters applied in percentage and visualizaing next
client_type = pre_data['NAME_CLIENT_TYPE']
client_type.head()
```

```python
app_client=round((client_type.value_counts()/pre_data["NAME_CLIENT_TYPE"].count())*100,2)
app_client = pd.DataFrame(app_client)
app_client.reset_index(level=0, inplace=True)
app_client.rename(columns=  {"index": "NAME_CLIENT_TYPE", 
                     "NAME_CLIENT_TYPE":"Percentage"}, 
                                 inplace = True) 
app_client.sort_values(by = 'Percentage' , inplace = True, ascending = False)
app_client
```

```python
plt.bar(app_client['NAME_CLIENT_TYPE'],app_client['Percentage'],color=['green', 'red', 'blue', 'cyan'])
plt.show()
# 73.72% of repeaters here who reapplied for loan while 18.04 are new clients.
```

```python
loan_purpose= pre_data['NAME_CASH_LOAN_PURPOSE']
loan_purpose.head()
```

```python
loan_reason=round((loan_purpose.value_counts()/pre_data["NAME_CASH_LOAN_PURPOSE"].count())*100,2)
loan_reason = pd.DataFrame(loan_reason)
loan_reason.reset_index(level=0, inplace=True)
loan_reason.rename(columns=  {"index": "NAME_CASH_LOAN_PURPOSE", 
                     "NAME_CASH_LOAN_PURPOSE":"Percentage"}, 
                                 inplace = True) 
loan_reason.sort_values(by = 'Percentage' , inplace = True, ascending = False)
loan_reason
```

```python
plt.bar(loan_reason['NAME_CASH_LOAN_PURPOSE'],loan_reason['Percentage'],color=['green', 'red', 'blue', 'cyan'])
plt.xticks(rotation = 90)
plt.show()
#XAP is the most reason of the loan for the clients
```

```python
weekday_app_start= pre_data['WEEKDAY_APPR_PROCESS_START']
weekday_app_start.head()
```

```python
best_loan_day=round((weekday_app_start.value_counts()/pre_data["WEEKDAY_APPR_PROCESS_START"].count())*100,2)
best_loan_day = pd.DataFrame(best_loan_day)
best_loan_day.reset_index(level=0, inplace=True)
best_loan_day.rename(columns=  {"index": "WEEKDAY_APPR_PROCESS_START", 
                     "WEEKDAY_APPR_PROCESS_START":"Percentage"}, 
                                 inplace = True) 
best_loan_day.sort_values(by = 'Percentage' , inplace = True, ascending = False)
best_loan_day
```

```python
plt.bar(best_loan_day['WEEKDAY_APPR_PROCESS_START'],best_loan_day['Percentage'],color=['green', 'red','yellow', 'blue', 'cyan'])
plt.xticks(rotation = 45)
plt.show()
#Clients prefer to take or apply loans preferably on tuesdays and wednesdays
```
